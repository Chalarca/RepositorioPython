### Operaciones con Series

es importante tener en cuenta los tamaños de las series

```python
serie_3 = pd.Series(data = [1,2,3,4])
serie_4 = pd.Series(data = [100,200,300,400])
```

```python
serie_3+serie_4
```

```python
resultado=serie_3**serie_3
resultado
```

```python
#saldra un NaN si no se puede realizar alguna operacion
ahre=pd.Series([1,2,3])
resultado-ahre
```
