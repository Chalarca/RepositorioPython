```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.datasets import mnist
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import cv2


#model = tf.keras.models.load_model('D:\Proyectos Visual Studio\Python-Jupyther\Fund_Ap_Maq_Vision\Modelo80.h5')

model = tf.keras.models.load_model('D:\Proyectos Visual Studio\Python-Jupyther\Fund_Ap_Maq_Vision\my_model.h5')
```

```python
imageninput=cv2.imread("D:\\Proyectos Visual Studio\Python-Jupyther\Fund_Ap_Maq_Vision\\Numeros.jpg",0)
_,imageninput=cv2.threshold(imageninput, 80, 255, cv2.THRESH_BINARY)
imagen=imageninput/255
h,w=imagen.shape[:2]
imagen=cv2.resize(imagen,[h//2,w//2])
imagenshow=imagen.copy()
h,w=imagen.shape[:2]
tam_vent=150

for i in range(0,h-tam_vent,50):
    j=0
    if i+tam_vent >h:
        i=h-tam_vent
    else:
        0

    for j in range(0,w-tam_vent,50):

        if j+tam_vent > w:
            j=w-tam_vent
        else:
            0
        ventana=imagen[i:tam_vent+i,j:tam_vent+j]
        datos1=cv2.resize(ventana,[28,28])
        #print(datos1.sum())
        datos=np.reshape(datos1,(1,28,28,1))
        if datos1.sum() <760:
            salida=model.predict(datos,verbose=0)
        
            print(salida.argmax(),salida.max())

            #print(tam_vent+i,tam_vent+j)

            #cv2.imshow("numero", ventana)
            #cv2.waitKey(0) 
            #cv2.destroyAllWindows()
            if salida.max()>=0.96:
                cv2.rectangle(imagenshow,(j,i),(tam_vent+j,tam_vent+i),(0,255,0),2)
                cv2.putText(imagenshow,str(salida.argmax()),(j, i+25),cv2.FONT_HERSHEY_SIMPLEX, 1, (0,255,0), 1, cv2.LINE_AA)
        #datos=cv2.cvtColor(datos, cv2.COLOR_BGR2GRAY)
        
        #print(datos.shape)

      
    #print(salida)

#Hago la ventana deslizante
#cv2.imshow("numero", ventana)
cv2.imshow("imagen", imagenshow)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```
