```python
from scipy.io import wavfile
import matplotlib.pyplot as plt
import scipy.io as sio
from scipy.fftpack import fft
import numpy as np
```

```python
fs,Audiodata = wavfile.read("D:\\Users\ACER\Desktop\Trabajo Investigacion\Aureas Mono especies\Audio1\JAGUAS253_20121114_080107.wav")
AudiodataScaled = Audiodata[:,0]/(2**15)

#definir los valores del eje x en milisegundos
timeValues = np.arange(0, len(AudiodataScaled), 1)/ fs # Convertir Muestras/Seg a Segundos
timeValues = timeValues * 1000  #Escala de tiempo en milisegundos
plt.rcParams['figure.figsize'] = (15, 7)
plt.plot(timeValues, AudiodataScaled);plt.title('Audio Analisado',size=16)
plt.ylabel('Amplitud'); plt.xlabel('Tiempo (ms)');
```

```python
fs,Audiodata = wavfile.read("D:\\Users\ACER\Desktop\Trabajo Investigacion\Aureas Mono especies\Audio1\JAGUAS253_20121114_080107.wav")
Audiodata=Audiodata[:,0]
Audiodata = Audiodata
n = len(Audiodata) 
AudioFreq = fft(Audiodata)
MagFreq = np.abs(AudioFreq)
MagFreq = MagFreq / float(n)

plt.plot(MagFreq) #Espectro de magnitud
plt.ylabel('Magnitud',fontsize = 16); plt.title('FFT total',fontsize = 16)
```
